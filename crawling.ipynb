```python
from openpyxl import Workbook

url = 'https://www.kobis.or.kr/kobis/business/mast/mvie/searchMovieList.do' 

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36"}
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument("--single-process")
chrome_options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Chrome('/usr/bin/chromedriver', options=chrome_options)
driver.get(url)

with open("/home/ubuntu/youngmi/synopsis_data.xlsx", "w", encoding='utf-8') as f:
    wb = Workbook()
    ws = wb.active
    ws.append(["title","code","synopsis"])

    html = driver.page_source #URL에 해당하는 페이지의 HTML를 가져옴
    soup1 = BeautifulSoup(driver.page_source, 'html.parser')
    table = soup1.select('table > tbody > tr')
        # 현재 8412페이지 까지 
    for page in range(1,3):
        driver.execute_script(f"goPage('{page}')") # 페이지 이동
    # content > div.rst_sch > table > tbody > tr:nth-child(1) > td:nth-child(1) > span > a
        for i in range(len(table)):
            title = table[i].select_one('td:nth-child(1) > span').text.strip() # 영화 제목
            code = table[i].select_one('td:nth-child(3)> span').text.strip() # 영화 코드
            
            # 팝업  //*[@id="content"]/div[4]/table/tbody/tr[4]
            xpath = f'//*[@id="content"]/div[4]/table/tbody/tr[{i+1}]/td[1]/span/a'
            driver.find_element_by_xpath(xpath).click() # 영화 클릭
            synopsis = driver.find_element_by_xpath("//*[contains(text(),'시놉시스')]/following-sibling::p").text.strip()
            driver.execute_script("dtlRmAll();return false;") # 창 닫기

            row = [title, code, synopsis]
            ws.append(row) # 엑셀 파일에 데이터 추가
```
